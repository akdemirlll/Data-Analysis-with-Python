## Excel
Pandas can read and write excel files, keep in mind, this only imports data. Not formulas or images, having images or macros may cause this read_excel method to crash. 

### Excel Input

```python
df = pd.read_excel("Excel_Sample.xlsx")
```

```python
df
```

```python
pd.read_excel("Excel_Sample.xlsx", sheet_name = "Sheet2")
```

```python
df1 = pd.read_excel("ornekx.xlsx")
```

```python
df1
```

### Excel Output

```python
df.to_excel("Excel_Sample1.xlsx", sheet_name = "Sheet1", index = False)
```

```python
df1.to_excel("Excel_Sample2.xlsx", sheet_name = "Sheet2", index = False)
```
